```python
from utils import process_input

def line_fn(line):
    ranges = line.split(',')
    output = []
    for range in ranges:
        lb,ub = range.split('-')
        output.append((int(lb), int(ub)))
    return output
    

sample = process_input("sample.txt", line_fn)[0]
full = process_input("full.txt", line_fn)[0] 
```

```python
def only_repeating(s):
    if len(s) % 2 != 0: # Only even sized 
        return False
    
    i = len(s) // 2

    if s[0:i] != s[i:]:
        return False
    return True


def count_invalid_in_range(ub,lb):
    return sum(filter(lambda x: only_repeating(str(x)), range(ub, lb+1)))

def part_1(input):
    return sum([count_invalid_in_range(ub,lb) for ub, lb in input])
```

```python
part_1(sample)
```

```python
part_1(full)
```

```python
def only_repeating_subseq(s):
    for i in range(1,(len(s) // 2)+1):
        check_pattern = s[:i]
        check_string = check_pattern * (len(s) // i)
        #print(f"{i}: {check_pattern} * ({len(s)} // {i})-> {check_string}")
        if check_string == s:
            return True
    return False

def count_invalid_part_2(ub,lb):
    return sum(filter(lambda x: only_repeating_subseq(str(x)), range(ub, lb+1)))

def part_2(input):
    return sum([count_invalid_part_2(ub,lb) for ub, lb in input])
```

```python
part_2(sample)
```

```python
part_2(full)
```

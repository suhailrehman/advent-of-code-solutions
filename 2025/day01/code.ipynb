```python
from utils import process_input

def extract_lock(line: str) -> tuple[str,int]:
    return line[0], int(line[1:])


input = process_input("sample.txt", extract_lock)
full = process_input("full.txt", extract_lock) 
```

```python
def part_1(input):
    pos = 50
    zero_count = 0
    for dir, num in input:
        num = num % 100
        #print(f"Current Pos: {pos}")
        if dir == "L":
            pos = (pos - num)
        else:
            pos = (pos + num)
        if pos < 0:
            pos = 100 + pos
        elif pos > 99:
            pos = pos - 100
        
        if pos == 0:
            zero_count += 1
        print(f"Rotated {dir}{num} to {pos}")
    return zero_count
```

```python
part_1(input)
```

```python
def part_2(input, start=50):
    pos = start
    zero_count = 0
    for dir, num in input:
        if dir == "L":
            pos = (pos - num) 
        else:
            pos = (pos + num)

        if pos > 100 or pos < 0:
            zero_count += abs(pos // 100)

        pos = pos % 100
    
    return zero_count
```

```python
def part_2(input, start=50):
    pos = start
    zero_count = 0
    d = {"L": -1, "R": +1}
    for dir, num in input:
        for _ in range(num):
            pos += d[dir]
            if pos % 100 == 0:
                zero_count +=1

    return zero_count
```

```python
def compute_successor(l):
    if l == []:
        return 0
    if len(set(l)) == 1:
        return l[-1]
    
    else:
        return l[-1] + compute_successor([l[i+1] - l[i] for i in range(len(l)-1)])
    

def part_one(input):
    results = []

    for line in input:
        results.append(compute_successor(line))

    return sum(results)


    
```

```python
def compute_predecesor(l):
    #print(l)
    if l == []:
        return 0
    if len(set(l)) == 1:
        return l[0]
    
    else:
        return l[0] - compute_predecesor([l[i+1] - l[i] for i in range(len(l)-1)])
    

def part_two(input):
    results = []

    for line in input:
        results.append(compute_predecesor(line))

    return sum(results)
```

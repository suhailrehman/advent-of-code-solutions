```python
from parse import parse

def parse_input(input):
    map = {}
    directions = []
    start_pos = None
    end_pos = None
    for line in input:
       
        
        if line.strip() == '':
            continue
        elif '=' in line:
            # Use the parse library to parse AAA = (BBB,CCC) into three varaibles
            source, left, right = parse("{} = ({}, {})", line.strip())
            if not start_pos:
                start_pos = source
            #print(source, left, right) 
            map[source] = (left, right)
            end_pos = source
        else:
            directions = list(line.strip())

    return directions, map, start_pos, end_pos
```

```python
directions, map, start_pos, end_pos = parse_input(sample_input)
```

```python
def part_one(directions, map, start_pos, end_pos):
    loc = start_pos
    steps = 0
    while True:
        for d in directions:
            #print(d, loc)
            left, right = map[loc]
            if d == 'L':
                loc = left
            elif d == 'R':
                loc = right
            else:
                raise Exception('Unknown direction')
            
            steps += 1
            if loc == end_pos:
                return steps

        
    
```

```python
directions, map, start_pos, end_pos = parse_input(sample_input)
#print(start_pos,end_pos)
print(part_one(directions, map, start_pos, end_pos))
```

```python
directions, map, start_pos, end_pos = parse_input(full_input)
print(part_one(directions, map, 'AAA', 'ZZZ'))
```

```python
def part_two(directions, map):
    locs = [x for x in map.keys() if x[-1] == 'A']
    steps =0
    print(f"There are {locs} starting locations")
    while True:
        for d in directions:
            #print(d, locs)
            locs_copy = locs[:]
            for ix, loc in enumerate(locs):
                left, right = map[loc]
                if d == 'L':
                    locs_copy[ix] = left
                elif d == 'R':
                    locs_copy[ix] = right
                else:
                    raise Exception('Unknown direction')
                
            locs = locs_copy
            steps += 1
            
            if all([x[-1] == 'Z' for x in locs_copy]):
                return steps
```

```python
directions, map, start_pos, end_pos = parse_input(sample_input)
print(part_two(directions, map))
```

```python
directions, map, start_pos, end_pos = parse_input(full_input)
print(part_two(directions, map))
```

```python
def part_two_lcm(directions, map):
    locs = [x for x in map.keys() if x[-1] == 'A']
    print(f"There are {locs} starting locations")
    all_steps = []

    for loc in locs:
        steps = 0
        stop = False
        while stop == False:
            for d in directions:
                #print(d, loc)
                left, right = map[loc]
                if d == 'L':
                    loc = left
                elif d == 'R':
                    loc = right
                else:
                    raise Exception('Unknown direction')
                
                steps += 1
                if loc.endswith('Z'):
                    all_steps.append(steps)
                    stop = True
                    break
    
    print(all_steps)

    return all_steps
```

```python
directions, map, start_pos, end_pos = parse_input(full_input)
all_steps = part_two_lcm(directions, map)
```

```python
from math import lcm

print(lcm(*all_steps))
```

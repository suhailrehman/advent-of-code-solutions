```python
def build_maps(input):
    seeds = []
    maps = {}
    map_type = None
    current_map = {}
    for line in input:
        if line.startswith('seeds:'):
            seeds = [int(x) for x in line.split(':')[1].strip().split()]
        elif 'map:' in line:
            map_type = tuple(line.split('map:')[0].strip().split('-to-'))
            maps[map_type] = []
        elif line.strip() == []:
            continue
        elif map_type and line.strip():
            #print(line, map_type)
            maps[map_type].append([int(x) for x in line.strip().split()])
    
    return seeds, maps


```

```python
steps = ['seed', 'soil', 'fertilizer', 'water', 'light', 'temperature', 'humidity', 'location']

# get all linear two sequences of steps here
step_sequences = [(steps[i], steps[i+1]) for i in range(len(steps)-1)]
step_sequences

def part_one(seeds, maps):
    locations = []
    for seed in seeds:
        point = seed
        for ix, (step1, step2) in enumerate(step_sequences):
            mapping_found = False
            for dest, source, length in maps[(step1, step2)]:
                if source <= point <= source + length:
                    #print(f"{step1}->{step2}: {point} -> {dest + (point - source)}")
                    map_ix = point - source
                    point = dest + map_ix
                    mapping_found = True
                    break
            if not mapping_found:
                #print(f"no mapping found for {step1}->{step2}: {point}")
                pass

            if step2 == 'location':
                locations.append(point)
        
    return locations
            
       


        
    
```

```python
rev_steps = ['location',
 'humidity',
 'temperature',
 'light',
 'water',
 'fertilizer',
 'soil',
 'seed']

# get all linear two sequences of steps here
step_sequences = [(rev_steps[i], rev_steps[i+1]) for i in range(len(rev_steps)-1)]
step_sequences

def generate_seeds(seeds):
    for ix in range(0,len(seeds),2):
        seed_range_start = seeds[ix]
        num_seeds = seeds[ix+1]
        for seed in range(seed_range_start,seed_range_start+num_seeds):
            #print(seed)
            yield seed
        #print('next range')
    

def find_location(seed, maps):
    point = seed
    for ix, (step1, step2) in enumerate(step_sequences):
        mapping_found = False
        for dest, source, length in maps[(step1, step2)]:
            if source <= point <= source + length:
                #print(f"{step1}->{step2}: {point} -> {dest + (point - source)}")
                map_ix = point - source
                point = dest + map_ix
                mapping_found = True
                break
        if not mapping_found:
            #print(f"no mapping found for {step1}->{step2}: {point}")
            pass

        if step2 == 'location':
            return point
        
def find_seed(location, maps):
    point = location
    for ix, (step2, step1) in enumerate(step_sequences):
        mapping_found = False
        for source, dest, length in maps[(step1, step2)]:
            if source <= point <= source + length:
                #print(f"{step1}->{step2}: {point} -> {dest + (point - source)}")
                map_ix = point - source
                point = dest + map_ix
                mapping_found = True
                break
        if not mapping_found:
            #print(f"no mapping found for {step1}->{step2}: {point}")
            pass

        if step1 == 'seed':
            return point
        

def seed_intervals(seeds):
    for ix in range(0,len(seeds),2):
        seed_range_start = seeds[ix]
        num_seeds = seeds[ix+1]
        yield seed_range_start, seed_range_start+num_seeds

def part_two(seeds, maps):
    for location in range(0, 100000000):
        seed = find_seed(location, maps)
        # Lookup seed in seeds
        #print(location,seed)
        for seed_range_start, seed_range_end in seed_intervals(seeds):
            if seed in range(seed_range_start, seed_range_end):
                return location

```

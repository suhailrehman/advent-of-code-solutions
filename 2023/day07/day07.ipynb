```python
from collections import Counter
def parse_input(input):
    hands = []
    for line in input:
        #hand_count = Counter()
        hand, bid = line.strip().split()
        #print(hand, bid)
        #hand_count.update(hand)
        #hands.append((hand_count, int(bid)))
        hands.append((hand, int(bid)))

    return hands


```

```python
from collections import defaultdict
from functools import cmp_to_key

ranks = ['J', '2','3','4','5','6','7','8','9','T','Q','K','A']
ranks_one = ['2','3','4','5','6','7','8','9','T', 'J', 'Q','K','A']


# def handount):
#     highest_rank = 0
#     for r, c in hand_count.items():
#         if c == count:
#             if ranks.index(r) > highest_rank:
#                 highest_rank = ranks.index(r)
#     return highest_rank

def eval_hand(hand, part_two=False):
    if 'J' in hand and part_two:
        try:
            hand_copy = hand[:]
            hand_count = Counter(hand)
            if hand_count['J'] == 5:
                return 7, hand
            
            #most_common_count = hand_count.most_common()[0]
            #most_common_ranks = [r for r, c in hand_count.items() if c == most_common_count[1] and r != 'J']
            most_common_ranks = []
            for r, c in hand_count.most_common():
                if r == 'J':
                    continue
                most_common_ranks.append(r)
                break
            
            substitute_card = sorted(most_common_ranks, key=lambda x: ranks.index(x), reverse=True)[0]
            print(hand_copy, substitute_card)
            hand_copy = hand_copy.replace('J', substitute_card)
            hand_count = Counter(hand_copy)
        except IndexError:
            print('IE:', hand, hand_count)
            raise

    else:
        hand_count = Counter(hand)

    if 5 in hand_count.values():
        return 7, hand
    elif 4 in hand_count.values():
        return 6, hand
    elif 3 in hand_count.values() and 2 in hand_count.values():
        return 5, hand
    elif 3 in hand_count.values():
        return 4, hand
    elif 2 in hand_count.values() and len(hand_count) == 3:
        return 3, hand
    elif 2 in hand_count.values():
        return 2, hand
    else:
        return 1, hand

def compare_hands(hand_bid1, hand_bid2):
    hand1 = hand_bid1[0]
    hand2 = hand_bid2[0]
    for card1, card2 in zip(hand1, hand2):
        if ranks.index(card1) > ranks.index(card2):
            return -1
        elif ranks.index(card1) < ranks.index(card2):
            return 1
        
    return 0
    
def part_one(hands, part_two=False):
    hand_ranks = defaultdict(list)
    for hand, bid in hands:
        rank, hand = eval_hand(hand,part_two=part_two)
        hand_ranks[rank].append((hand, bid))


    for k,v in hand_ranks.items():
        hand_ranks[k] = sorted(v, key=cmp_to_key(compare_hands), reverse=True)

    #print(hand_ranks)

    hand_total = 0
    count_rank = 1
    for hand_rank in sorted(hand_ranks.keys()):
        for hand, bid in hand_ranks[hand_rank]:
            print(hand, bid, count_rank, bid * count_rank)
            hand_total += bid * count_rank
            count_rank += 1

    return hand_total
```

```python
# Line Input
sample_input = read_input('sample.txt')
full_input = read_input('input.txt')

```

```python
def parse_input(input):
    for line in input:
        if line.startswith('Time'):
            times = [int(x) for x in line.split(':')[-1].split()]
        elif line.startswith('Distance'):
            distances = [int(x) for x in line.split(':')[-1].split()]

    return {x: y for x,y in zip(times, distances)}





```

```python
race_times = parse_input(sample_input)
race_times
```

```python
import math

def part_one(race_times):
    combinations = []
    for time, record in race_times.items():
        winning_times = 0
        for hold_time in range(1,time):
            race_time = time - hold_time
            speed = hold_time
            distance = speed * race_time

            if distance > record:
                winning_times += 1
        combinations.append(winning_times)

    return combinations
    
```

```python
part_one(race_times)
```

```python
race_times = parse_input(full_input)
math.prod(part_one(race_times))
```

```python
def parse_input_two(input):
    for line in input:
        if line.startswith('Time'):
            times = int(''.join([x for x in line.split(':')[-1].split()]))
        elif line.startswith('Distance'):
            distances = int(''.join([x for x in line.split(':')[-1].split()]))

    return {times: distances}

    
```

```python
race_times = parse_input_two(sample_input)
math.prod(part_one(race_times))
```

```python
race_times = parse_input_two(full_input)
math.prod(part_one(race_times))
```

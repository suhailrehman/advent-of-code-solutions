```python
import numpy as np

s = len("XMAS")

def strfy(lst):
    return ''.join(lst)

def count_xmas(matrix, i, j):
    surrounds = []
    x,y = len(matrix), len(matrix[0])
    
    if j + s <= y:  # Check row forward
        surrounds.append(strfy(matrix[i][j+k] for k in range(s)))
    if j - s + 1 >= 0:  # Check row back
        surrounds.append(strfy(matrix[i][j-k] for k in range(s)))
    if i + s <= x: # Check column down
        surrounds.append(strfy([matrix[i+k][j] for k in range(s)]))
    if i - s + 1 >=0: # Check column up
        surrounds.append(strfy([matrix[i-k][j] for k in range(s)]))
    
    #Diagonals
    if j + s <= y and i + s <= x: # SE diagonal
        surrounds.append(strfy([matrix[i+k][j+k] for k in range(s)]))
    if j + s <= y and i - s + 1>=0: # NE diagonal
        surrounds.append(strfy([matrix[i-k][j+k] for k in range(s)]))
    if j - s + 1 >= 0 and i - s + 1 >=0: # NW diagonal
        surrounds.append(strfy([matrix[i-k][j-k] for k in range(s)]))
    if j - s + 1>=0 and i + s <= x: # SW diagonal
        surrounds.append(strfy([matrix[i+k][j-k] for k in range(s)]))

    return surrounds


def count_mas(matrix, i, j):
    # bounds check
    x,y = len(matrix), len(matrix[0])
    if i - 1 < 0 or i + 1 >= x or j -1 < 0 or j+1 >= y:
        return False 
    
    diag1 = [matrix[i-1][j-1], matrix[i][j], matrix[i+1][j+1]]
    diag2 = [matrix[i+1][j-1], matrix[i][j], matrix[i-1][j+1]]

    print(i,j, diag1, diag2)

    if (strfy(diag1) == "MAS" or strfy(diag1[::-1]) == "MAS") and (strfy(diag2) == "MAS" or strfy(diag2[::-1]) == "MAS"):
        return True

    # if matrix[i-1][j-1] == 'M' and \
    #     matrix[i-1][j+1] == 'S' and \
    #     matrix[i][j] == 'A' and \
    #     matrix[i+1][j-1] == 'M' and \
    #     matrix[i+1][j+1] == 'S':
    #         return True
    
    return False

def part1(filename):
    matrix = []
    with open(filename) as fp:
        for line in fp:
            matrix.append(list(line.strip()))

    xmases = []
    for i, row in enumerate(matrix):
        for j, val in enumerate(row):
            if val == "X":
                #print(i,j,count_xmas(matrix,i,j))
                xmases.extend((count_xmas(matrix, i,j)))

    xmases = np.array(xmases)
    
    return xmases[xmases == "XMAS"].size

def part2(filename):
    matrix = []
    with open(filename) as fp:
        for line in fp:
            matrix.append(list(line.strip()))

    mases = []
    for i, row in enumerate(matrix):
        for j, val in enumerate(row):
            if val == "A":
                print(i,j,count_mas(matrix,i,j))
                mases.append((count_mas(matrix, i,j)))

    
    
    return sum(mases)

                          
```

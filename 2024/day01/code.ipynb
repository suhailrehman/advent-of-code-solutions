```python
import numpy as np

def part1(filename):
    data = np.loadtxt(filename)
    data.sort(axis=0)
    return np.sum(np.abs(data[:,0] - data[:,1]))

def part2(filename):
    data = np.loadtxt(filename)
    col1 = data[:,0]
    col2 = data[:,1]

    unique, counts = np.unique(col2, return_counts=True)
    count_dict = dict(zip(unique, counts))
    return np.sum([num * count_dict.get(num, 0) for num in col1])
```

```python
part1("sample.txt"), part1("full.txt")
```

```python
part2("sample.txt"), part2("full.txt")
```
